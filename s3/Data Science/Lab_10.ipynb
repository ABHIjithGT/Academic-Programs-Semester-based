# **1. Iris Dataset**

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from google.colab import files

# Upload iris dataset CSV
uploaded = files.upload()

data = pd.read_csv('iris.csv')

X = data.drop('Species', axis=1)
y = data['Species']

def run_gaussian_nb(test_size):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=42)
    model = GaussianNB()
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    print(f"Train-Test Split: {1-test_size:.0%} - {test_size:.0%}")
    print("Classification Report:")
    print(classification_report(y_test, y_pred))
    print("Confusion Matrix:")
    print(confusion_matrix(y_test, y_pred))
    print(f"Accuracy: {accuracy_score(y_test, y_pred):.4f}")
    print("-"*40)

print("Iris Dataset Results\n")
run_gaussian_nb(0.3)
run_gaussian_nb(0.2)
```

# **2. Breast Cancer Dataset**

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from google.colab import files

# Upload dataset
uploaded = files.upload()
data = pd.read_csv('Breast_Cancer.csv')

# Drop irrelevant columns
data = data.drop(['id', 'Unnamed: 32'], axis=1, errors='ignore')

# Fill missing values only in numeric columns
numeric_cols = data.select_dtypes(include=['float64', 'int64']).columns
data[numeric_cols] = data[numeric_cols].fillna(data[numeric_cols].mean())

# Encode the target column 'diagnosis' (convert 'M'/'B' to 1/0)
le = LabelEncoder()
data['diagnosis'] = le.fit_transform(data['diagnosis'])

# Split features and target
X = data.drop('diagnosis', axis=1)
y = data['diagnosis']

print(f"Features shape: {X.shape}, Target shape: {y.shape}")

def run_gaussian_nb(test_size):
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=test_size, random_state=42)
    model = GaussianNB()
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    print(f"Train-Test Split: {1-test_size:.0%} - {test_size:.0%}")
    print("Classification Report:")
    print(classification_report(y_test, y_pred))
    print("Confusion Matrix:")
    print(confusion_matrix(y_test, y_pred))
    print(f"Accuracy: {accuracy_score(y_test, y_pred):.4f}")
    print("-"*40)

print("Breast Cancer Dataset Results\n")
run_gaussian_nb(0.3)
run_gaussian_nb(0.2)
```

# **4. Wine Quality Dataset**

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.preprocessing import LabelEncoder
from google.colab import files

uploaded = files.upload()
data = pd.read_csv('Wine_Quality.csv')

print(data.columns)
print(data.dtypes)

# Handle missing values
numeric_cols = data.select_dtypes(include=['float64', 'int64']).columns
data[numeric_cols] = data[numeric_cols].fillna(data[numeric_cols].mean())

categorical_cols = data.select_dtypes(include=['object']).columns
for col in categorical_cols:
    data[col] = data[col].fillna(data[col].mode()[0])

# Encode categorical columns
le = LabelEncoder()
for col in categorical_cols:
    data[col] = le.fit_transform(data[col])

# Define target and features
target_col = 'quality'
X = data.drop(target_col, axis=1)
y = data[target_col]

def run_gaussian_nb(test_size):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=42)
    model = GaussianNB()
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    print(f"Train-Test Split: {1-test_size:.0%} - {test_size:.0%}")
    print("Classification Report:")
    print(classification_report(y_test, y_pred))
    print("Confusion Matrix:")
    print(confusion_matrix(y_test, y_pred))
    print(f"Accuracy: {accuracy_score(y_test, y_pred):.4f}")
    print("-"*40)

print("Wine Quality Dataset Results\n")
run_gaussian_nb(0.3)
run_gaussian_nb(0.2)
```
