```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Data
age = np.array([18, 22, 30, 45, 65, 80]).reshape(-1, 1)
accident_no = np.array([38, 36, 24, 20, 18, 28])

# Create and fit linear regression model
model = LinearRegression()
model.fit(age, accident_no)

# Get slope and intercept
slope = model.coef_[0]
intercept = model.intercept_

print(f"Regression Line: accident_no = {slope:.2f} * age + {intercept:.2f}")

# Predict number of accidents for ages 40 and 60
ages_to_predict = np.array([[40], [60]])
predicted_accidents = model.predict(ages_to_predict)

for age_val, pred in zip(ages_to_predict.flatten(), predicted_accidents):
    print(f"Predicted accidents for age {age_val}: {pred:.2f}")

# Plotting
plt.scatter(age, accident_no, color='blue', label='Data Points')
plt.plot(age, model.predict(age), color='red', label='Regression Line')
plt.scatter(ages_to_predict, predicted_accidents, color='green', label='Predictions (40 & 60)')
plt.xlabel("Age")
plt.ylabel("Number of Accidents")
plt.title("Simple Linear Regression: Age vs Number of Accidents")
plt.legend()
plt.grid(True)
plt.show()
```
